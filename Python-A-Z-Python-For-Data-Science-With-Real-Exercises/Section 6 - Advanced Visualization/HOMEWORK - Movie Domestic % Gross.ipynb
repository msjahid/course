```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
movies = pd.read_csv('Section6-Homework-Data.csv', encoding='latin1')
```

```python
#Filter the dataframe by genre
#movies2 = movies[(movies.Genre == 'action') | (movies.Genre == 'comedy') | (movies.Genre == 'adventure') | \
#                 (movies.Genre == 'animation') | (movies2.Genre == 'drama')]
genre_filters = ['action', 'comedy', 'adventure', 'animation', 'drama']
movies2 = movies[movies.Genre.isin(genre_filters)]
```

```python
#Filter the dataframe by studio
#movies3 = movies2[(movies.Studio == 'Buena Vista Studios') | (movies.Studio == 'Sony') | \
#                  (movies.Studio == 'Universal') | (movies.Studio == 'WB') | \
#                  (movies.Studio == 'Paramount Pictures') | (movies.Studio == 'Fox')]
studio_filters = ['Buena Vista Studios', 'Sony', 'Universal', 'WB', 'Paramount Pictures', 'Fox']
movies3 = movies2[movies.Studio.isin(studio_filters)]
```

```python
#Define the style
sns.set(style='darkgrid', palette='muted', color_codes=True)

#Create subplots to control figure size
fig, ax = plt.subplots(figsize=(8.7, 6.27))

#Plot boxplots
ax = sns.boxplot(data=movies3, x='Genre', y='Gross % US', color='lightgrey')
#Set transparency
plt.setp(ax.artists, alpha=0.5)

#Add points
sns.stripplot(data=movies3, x='Genre', y='Gross % US', hue='Studio', jitter=True, size=6)

#Position of the legend
plt.legend(loc='upper left', bbox_to_anchor=(1,1), fontsize=13)
#ax.legend(loc='upper left', bbox_to_anchor=(1,1), fontsize=13)

#Set title and x and y labels
ax.set_title('Domestic Gross % by Genre', fontsize=30)
#plt.title('Domestic Gross % by Genre', fontsize=30)
ax.set_xlabel('Genre', fontsize=20)
#plt.xlabel('Genre', fontsize=20)
ax.set_ylabel('Gross % US', fontsize=20)
#plt.ylabel('Gross % US', fontsize=20)
plt.xticks(fontsize=12)
plt.yticks(range(0,100,20), fontsize=12)
plt.legend()

plt.show()
```
